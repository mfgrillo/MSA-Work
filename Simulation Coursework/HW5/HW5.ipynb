```python
import numpy as np
import bisect
import matplotlib.pyplot as plt
```

**2a**

```python
# Create intervals for m
m = np.arange(0, 1.01, (1/10))

# List of discrete probabilities
discrete_probs = [.05, .09, .12, .08, .03, .16, .07, .22, .04, .14]

# Create list for cumulative distribution
cum_probs = np.cumsum(discrete_probs)

# Assign cutpoints

cutpoints = []

for cutpoint in cum_probs:
    cutpoints.append(bisect.bisect_left(m, cutpoint))

print(cutpoints)
```

**2b**

```python
# Create sample
num = 10000
sample = np.random.choice(np.arange(1,11), num, p = discrete_probs)

# Calculate the empirical CDF of the sample
unique_samples, sample_counts = np.unique(sample, return_counts=True)
sample_cumulative_probabilities = np.cumsum(sample_counts) / num

# P-P plot
plt.plot(cum_probs, sample_cumulative_probabilities, 'o', label='Frequencies')

# Reference line
plt.plot([0, 1], [0, 1], 'k--', label='Ideal 1:1 Line')

# Labels for margins and title
plt.xlabel('Theoretical CDF')
plt.ylabel('Empirical CDF')
plt.title('P-P Plot, Theoretical vs Empirical CDF')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.grid()

# Add legend
plt.legend()

# Show the plot
plt.show()
```

```python
# density function of logistic distribution

def logistic_grapher(X, gamma, beta):
    return ((1/beta)*np.exp(-(X-gamma)/beta))/\
            (1+np.exp(-(X-gamma)/beta))**2


# Sample 10000 values from U~(0,100)

x = np.random.uniform(0,100,10000)

# beta and gamma at 10

y = logistic_grapher(x,10,10)

# plot

plt.scatter(x, y)

plt.xlabel('x')

plt.ylabel('y')

plt.title('gamma = 10, beta = 10')

plt.show()

# gamma = 30, beta = 10

y = logistic_grapher(x,30,10)

# plot
plt.scatter(x, y)

plt.xlabel('x')

plt.ylabel('y')

plt.title('gamma = 30, beta = 10')

plt.show()

# gamma = 30, beta = 10

y = logistic_grapher(x,10,30)

plt.scatter(x, y)

plt.xlabel('x')

plt.ylabel('y')

plt.title('gamma = 10, beta = 30')

plt.show()
```

The above cell shows that, when gamma increases, the location of the peak changes, wheras when beta increases the scale or spread of the distribution. Thus, we can see that gamma and beta are the location and scale parameters respectively.
