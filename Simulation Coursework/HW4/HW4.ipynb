```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import kstest
from scipy.stats import norm
import math
import seaborn as sns
```

```python
# X value and peak from previous part
x_value = 8.8125
peak = 0.015

# Calculation of standard deviation
std_dev = 1 / (peak * np.sqrt(2 * np.pi))

# Number of samples
n_samples = 1000

# Random sample generation with and without mirroring
samples = np.absolute(np.random.normal(loc=x_value, scale=std_dev, size=n_samples))
samplesneg = np.random.normal(loc=x_value, scale=std_dev, size=n_samples)

# Histogram 1
plt.hist(samples, bins = 25)
plt.xlabel('X values')
plt.ylabel('Frequency')
plt.title('Histogram of Kernel samples using mirroring')
plt.show()

# Histogram 2
plt.hist(samplesneg, bins = 25)
plt.xlabel('X values')
plt.ylabel('Frequency')
plt.title('Histogram of Kernel samples with no mirroring')
plt.show()
```

```python
# Given parameters
gam = 9
mode = 22
ninetyfifth = 30

# Value of 95th percentile of a N(0, 1) distribution
norm_val = norm.ppf(0.95)

# calculation of c, sigma and mu
c = math.log((mode-gam)/(ninetyfifth-gam))
sigma = (-norm_val + math.sqrt(norm_val**2 - 4 * c))/2
mu = math.log(mode-gam) + sigma**2

# Generating 10k observations
test_array = np.random.lognormal(mu, sigma, 10000) + gam

# Plotting Density & CDF
seven = sns.displot(test_array, kind='hist', kde=True,
            bins=25, color='darkblue',
            edgecolor='black', linewidth=1)
seven.set(title='Density & CDF')

print("mu =", mu)
print("sigma =", mu)
```

```python
data = np.array([0.08, 0.1, 0.26, 0.35, 0.39, 0.24, 0.47, 0.11, 0.37, 0.09])

# cdf given
def EightCdensity(x):
    return 3*(1-x)**2

# Kolmogorov-Smirnov test
result = kstest(data, EightCdensity)

print("Test Statistic (D):", result.statistic)
print("P-value:", result.pvalue)

# Interpret the result
alpha = .1
if result.pvalue < alpha:
    print("Reject the null hypothesis: The data does not come from the specified distribution.")
else:
    print("Fail to reject the null hypothesis: The data may come from the specified distribution.")
```
